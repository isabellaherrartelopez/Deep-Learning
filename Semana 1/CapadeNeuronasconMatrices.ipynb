```
import numpy as np
```

**Procesamiento de un Neurona Artificial**

Crear el vector de entrada X

```
X=np.array([[1.0],[-0.5],[2.5],[0.5]])
print(X)
```

Crear vector de pesos W

```
W=np.array([[1.0],[-0.7],[-1.0],[0.5]])
print(W)
```

Calculo de la neta

```
Neta=(np.sum(np.multiply(X,W)))
print(Neta)
```

Function de activacion sigmoidal

```
Y=1.0/(1+np.exp(-Neta))
print(Y)
```

**Procesamiento de una capa de Neuronas**

W y B matricial, Wp(5x4), B(5x1) y X(4x1)

```
Wp=np.array([[1.5,0.5, -0.5,1.0],[0.7,-0.8, 1.5,-1.0],[-1.0,1.2, 0.3,-0.7],[1.0,0.2,-0.3,0.3],[-0.2,0.8,-1.5,0.6]])
print(Wp)
B=np.array([[1.0],[-0.5],[1.5],[0.7],[0.8]])
print(B)
X=np.array([[1.0],[-0.5],[2.5],[0.5]])
print(X)
```

Neta matricial

```
Neta=((np.matmul(Wp,X))+B)
print(Neta)
```

Salida Matricial

```
Y=1.0/(1+np.exp(-Neta))
print(Y)
```
